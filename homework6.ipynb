## 1. Modify String with Underscores
Given a string `txt`, insert an underscore (`_`) after every third character. If a character is a vowel or already has an underscore after it, shift the underscore placement to the next character. No underscore should be added at the end.

### Examples
**Input:** `hello`
**Output:** `hel_lo`

**Input:** `assalom`
**Output:** `ass_alom`

**Input:** `abcabcabcdeabcdefabcdefg`
**Output:** `abc_abcab_cdeabcd_efabcdef_g`


````python
# Assuming text1 is your input string
text1 = "hello" 
text2 = ('assalom')
text3 = ('abcabcabcdeabcdefabcdefg')

def modify_text(txt):
    vowels = "aeiouAEIOU"
    result = []
    count = 0
    
    for i, char in enumerate(txt):
        result.append(char)
        
        #No underscore should be added at the end
        if i == len(txt) - 1:
            break
            
        count += 1
        
        #Insert after every third character
        if count >= 3:
            #Character must not be a vowel AND not followed by an underscore
            if char not in vowels and txt[i+1] != '_':
                result.append('_')
                count = 0  # Success: Reset count
            else:
                #Shift placement to the next character
                count = 3  # Keep count high to try again at next char
                
    return "".join(result)

# Reassign text1 to the modified version
text1 = modify_text(text1)
text2 = modify_text(text2)
text3 = modify_text(text3)
#MUST print it to see the value in console
print(text1)
print(text2)
print(text3)
````

## 4. Return Uncommon Elements of Lists
### Task
Return the elements that are not common between two lists. The order of elements does not matter.

### Examples
- **Input:** `list1 = [1, 1, 2], list2 = [2, 3, 4]`  
  **Output:** `[1, 1, 3, 4]`

- **Input:** `list1 = [1, 2, 3], list2 = [4, 5, 6]`  
  **Output:** `[1, 2, 3, 4, 5, 6]`

- **Input:** `list1 = [1, 1, 2, 3, 4, 2], list2 = [1, 3, 4, 5]`  
  **Output:** `[2, 2, 5]`


````python
n1, n2 = 0, 1
count = 0

print("Fibanachi sequence:")
while count <10:
    print (n1, end = " ") 
    n1,n2 = n2, n1 + n2
    count +=1
````

````python
n = int(input())
fact = 1
for i in range(1, n+1):
    fact *=i
    print(f"{n}! =", fact)
````

````python
def uncommon(list1, list2):
    result = []
    for num in list1:
        if num not in list2:
            result.append(num)
    for num in list2:
        if num not in list1:
            result.append(num)
            return result
````
