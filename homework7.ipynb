## 1. is_prime(n) funksiyasi
`is_prime(n)` funksiyasini hosil qiling (`n > 0`). Agar `n` soni tub bo'lsa `True`, aks holda `False` qiymat qaytarsin.

### Misollar:
- **Kiritish:**  
  4  
  **Natija:**  
  False  
  _(Izoh: 4 soni tub emas, chunki u 2 ga bo'linadi.)_

- **Kiritish:**  
  7  
  **Natija:**  
  True  
  _(Izoh: 7 soni faqat 1 va o'ziga bo'linadi, ya'ni tub son.)_


```python
def is_prime(n):
    n < 0
    if n % 2 != 0 and n / 1 == n:
        return True 
    else:
       return False
is_prime(7)
```

## 2. digit_sum(k) funksiyasi
`digit_sum(k)` funksiyasini yozing, u `k` sonining raqamlari yig'indisini hisoblaydi.

### Misollar:
- **Kiritish:**  
  24  
  **Natija:**  
  6  
  _(Izoh: 24 sonining raqamlari yig'indisi: 2 + 4 = 6.)_

- **Kiritish:**  
  502  
  **Natija:**  
  7  
  _(Izoh: 502 sonining raqamlari yig'indisi: 5 + 0 + 2 = 7.)_


```python
digit_sum = lambda n: sum(int(digit) for digit in str(n))
print(digit_sum(502))
```

## 3. Ikki sonning darajalari
Berilgan `N` sonidan oshmaydigan barcha 2 ning darajalarini (ya'ni, `2**k` shaklidagi sonlarni) chop etuvchi funksiyani yozing.

### Misol:
- **Kiritish:**  
  10  
  **Natija:**  
  2 4 8  
  _(Izoh: 10 dan kichik yoki teng bo'lgan 2 ning darajalari: 2, 4, 8.)_

```python
powers = [2**i for i in range(10) if 2**i < 10]
print(powers)
```

-  In Python, functions are the "workhorses" of your code. You can group them into four main categories based on how they are created and used.

- 1. Built-in Functions.
Examples: print(), len(), type(), int(), sum().
Usage: length = len("Banking")


- 2. User-Defined Functions (UDFs)
These are the functions you create using the def keyword to perform specific tasks.

def welcome_user(name):
    return f"Hello, {name}!"

- 3. Anonymous Functions (Lambda Functions).

These are small, one-line functions that don't have a name. They are used for quick tasks, often inside other functions like filter() or map().

Syntax: lambda arguments: expression

Example: square = lambda x: x * x

- 4. Method Functions
Methods are functions that belong to a specific object (like a string or a list). You call them using a "dot" (.).

Examples: .append() for lists, .upper() for strings, or .execute() for database cursors.

Usage: "python".upper() results in "PYTHON".
