Homework 2. Simple Blog System

1. Define Post Class:
    - Create a Post class with attributes like title, content, and author.
2. Define Blog Class:
    - Create a Blog class that manages a list of posts.
    - Include methods to add a post, list all posts, and display posts by a specific author.
3. Create Main Program:
    - Develop a CLI to interact with the Blog system.
    - Include options to add posts, list all posts, and display posts by a specific author.
4. Enhance Blog System:
    - Add functionality to delete a post, edit a post, and display the latest posts.
5. Test the Application:
    - Create instances of posts and test the functionality of your Blog system.

```python
class Post:
    def __init__(self, title, content, author):#runs when we create a Post
        self.title = title
        self.content = content
        self.author = author
        
class Blog:
    def __init__(self):
        self.posts = []  #creates empty list
#It receives a Post object it adds that object to the list, Now Blog contains that post.
    def add_post(self, post):
        print(self.posts.append(post))#just add, no print
# It loops through all stored posts. Prints their title and author
    def list_post(self):
        for post in self.posts:
            print(post.title, "-", post.author)
#It checks each post. If author matches -> print title
    def get_author(self, author_name):
        for post in self.posts:
            if post.author == author_name:
                print(post.title)
blog = Blog()   #creating blog
#create posts
post1 = Post("Python", "Call", "Learning classes.")
post2 = Post("Math", "Uiliam", "Integral practice.")
#adds posts to blog
blog.add_post(post1)
blog.add_post(post2)

#show posts by specific author
blog.list_post()
blog.get_author("Call")
```

Homework 3. Simple Banking System

1. Define Account Class:
    - Create an Account class with attributes like account number, account holder name, and balance.
2. Define Bank Class:
    - Create a Bank class that manages a list of accounts.
    - Include methods to add an account, check balance, deposit money, and withdraw money.
3. Create Main Program:
    - Develop a CLI to interact with the Banking system.
    - Include options to add accounts, check balance, deposit money, and withdraw money.
4. Enhance Banking System:
    - Add functionality to transfer money between accounts, display account details, and handle account overdrafts.
5. Test the Application:
    - Create instances of accounts and test the functionality of your Banking system.

```python
class Account:
    def __init__(self, account_number, account_holder_name, balance = 0):
        self.account_number = account_number
        self.account_holder_name = account_holder_name
        self.balance = balance
        self.overdraft_limit = 100.0    #overdraft handling (limit set to $100)
#Enhance: Display account details.
    def display_details(self):
        """""Display account details."""
        print(f"\n---Account Details ---")
        print(f" Holder: {self.account_holder_name}")
        print(f"Number: {self.account_number}")
        print(f"Balance: ${self.overdraft_limit:.2f}")


class Bank:
    def __init__(self):
    #Manages an account.(Account number: Account object.)
        self.accounts = {}
    #Method to add an account.
    def add_account(self, account_number, holder, initial_deposit):
        if account_number not in self.accounts:

            new_account = Account(account_number, holder, initial_deposit)
            self.accounts[account_number] = new_account
            print(f"Account for {holder} created successufully.")
        else:
            print(f"Error: Account number already exists.")
#Method check balance.
    def get_balance(self, account_number, amount):
        acc= self.accounts.get(account_number)
        return acc.balance if acc else None
#Method deposit money.
    def deposit(self, account_number, amount):
        acc = self.accounts.get(account_number)
        if acc and amount > 0:
            acc.balance += amount
            return True
        return False
#Method to withdrow money(Enhanced with Overdraft.)
    def withdraw(self, account_number, amount):
        #Handle account overdrafts.
        acc = self.accounts.get(account_number)
        if acc:
        #Check if amount is within balance + overdraft limit
            if 0 < amount <+ (acc.balance + acc.overdraft_limit):
                acc.balance -+ amount
                return True
            return False
    
    def transfer(self, sender_num, receiver_num, amount):
        #Transfer money between accounts.
        sender = self.accounts.get(sender_num)
        receiver = self.accounts.get(receiver_num)

        if sender and receiver:
            if self.withdraw(sender_num, amount):
                self.deposit(receiver_num, amount)
                return True
            return False

#Create Main Program (CLI)
def main():
    my_bank = Bank()

    # 5. Test the Application: Pre-creating instances for testing
    my_bank.add_account("101", "Alice", 50.0)
    my_bank.add_account("102", "Bob", 20.0)
    print("Welcome! Test accounts '101' (Alice) and '102' (Bob) are ready.")

    while True:
        print("\n=== BANKING SYSTEM MENU ===")
        print("1. Open New Account")
        print("2. Check Balance & Details")
        print("3. Deposit Money")
        print("4. Withdraw Money")
        print("5. Transfer Money")
        print("6. Exit")
        
        choice = input("Select an option (1-6): ")

        try:
            if choice == '1':
                num = input("Enter unique Account Number: ")
                name = input("Enter Account Holder Name: ")
                bal = float(input("Enter Initial Deposit Amount: "))
                my_bank.add_account(num, name, bal)

            elif choice == '2':
                num = input("Enter Account Number: ")
                acc = my_bank.accounts.get(num)
                if acc:
                    acc.display_details()
                else:
                    print("Error: Account not found.")

            elif choice == '3':
                num = input("Enter Account Number: ")
                amt = float(input("Enter Deposit Amount: "))
                if my_bank.deposit(num, amt):
                    print(f"Success! New Balance: ${my_bank.get_balance(num):.2f}")
                else:
                    print("Error: Deposit failed.")

            elif choice == '4':
                num = input("Enter Account Number: ")
                amt = float(input("Enter Withdrawal Amount: "))
                if my_bank.withdraw(num, amt):
                    print(f"Success! New Balance: ${my_bank.get_balance(num):.2f}")
                else:
                    print("Error: Insufficient funds (overdraft limit reached).")

            elif choice == '5':
                s_num = input("Enter Your Account Number: ")
                r_num = input("Enter Recipient Account Number: ")
                amt = float(input("Enter Transfer Amount: "))
                if my_bank.transfer(s_num, r_num, amt):
                    print("Transfer successful!")
                else:
                    print("Error: Transfer failed. Check funds or account numbers.")

            elif choice == '6':
                print("Thank you for using the Banking System. Goodbye!")
                break
            else:
                print("Invalid choice. Please select 1-6.")
        
        except ValueError:
            print("Error: Please enter a valid numerical amount.")

if __name__ == "__main__":
    main()
```
