```python
# recursive function that calls itself to solve problem.
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

print (factorial(4))
```

```python
def fib_itarative(n):
    """"
calculates the nth Fibanacci num using an iterative approach
    """
    if n <= 1:
        return n
    a, b = 0, 1;
    for _ in range (2, n + 1):
        c = a + b
        a = b
        b = c 
        return b
result = fib_itarative(10)
print({result})
    
```

2. Define ToDoList Class:
    - Create a ToDoList class that manages a list of tasks.
    - Include methods to add a task, mark a task as complete, list all tasks, and display incomplete tasks.

```python
class To_Do():
    def __init__(self, add_task, complete_task, list_task, incomplete_task):
        self.add_task = add_task
        self.complete_task = complete_task
        self.list_task = list_task
        self.incomplete_task = incomplete_task
```

3. Create Main Program:
    - Develop a simple CLI to interact with the ToDoList.
    - Include options to add tasks, mark tasks as complete, list all tasks, and display only incomplete tasks.

```python
# """""
#         this program allows users to:
#         - add tasks
#         - mark tasks as complete
#         - view all tasks
#         - view only incomplete tasks
#         - exit the program
# """"

class To_Do():
    def __init__(self):
#Constructor method. Initializes an empty list to store taks.
        self.tasks = []
    def add_task(self, task): #Adds a new task to the task list.
        #parametrs: task(str): the description of the task.
        self.tasks.append({
            "task": task, "completed": False #By default, a new task is not completed
})
    def mark_complete(self, index):
#marks a task as completed based on its index.
#parametrs: index(int): position of the task in the list.
        if 0 <= index < len(self.tasks): #Check if index is valid
            self.tasks[index]["completed"] = True
        else:
            print("Invalid task number.")
            
    def list_tasks(self):
        if not self.tasks:
            print("No tasks available.")
            return
        for i, task in enumerate(self.tasks):
            status = "Done" if task["completed"] else "Not Done"
            print(f"{i}. {task["task"]} - {status}")
    def list_incomplete(self):
        found = False
        for i, task in enumerate(self.tasks):
            if not task["completed"]:
                print(f"{i}. {task["task"]} - Not Done")
                found = True
        if not found:
            print("All tasks are completed!")
      #MAIN PROGRAM (CLI Interface)      
if __name__ == "__main__":

#this block runs only when the file is executed directly. It creates a CLI (command line interface)
#to interact with ToDoList class.
    
    todo = To_Do() #create ToDo object
while True:
    #display meny options
    print("\n==== To Do List Menu ====")
    print("1. Add Task")
    print("2. Mark Task Complete")
    print("3. List All Tasks")
    print("4. Show Incomplete Tasks")
    print("5. Exit")

    choice = input("Choose option:")

    if choice == "1":
        task = input("Enter task description:")
        todo.add_task(task)
        print("Task added successfully.")
    elif choice == "2":
        try:
            index = int(input("Enter task number: "))
            todo.mark_complete(index)
        except ValueError:
            print("Please enter a valid number.")
    elif choice == "3":
        todo.list_tasks()
    elif choice == "4":
        todo.list_incomplete()
    elif choice == "5":
        print("Exiting program...")
        break
    #invalid input handling
    else:
        print("Invalid option. Please choose between 1 and 5.")
```

4. Test the Application:
    - Create instances of tasks and test the functionality of your ToDoList.

```python
todo = To_Do()
todo.add_task("Study Python")
todo.add_task("Drink more water")
todo.add_task("Read book")

todo.list_tasks()
todo.mark_complete(1)

print("\nAfter marking task 1 complete:")
todo.list_tasks()

print("\nIncomplete tasks only:")
todo.list_incomplete()
```
