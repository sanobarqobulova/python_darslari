```python
my_dict = {}

with open("example.txt","r")as example_file:
    for line in example_file:
        for word in line.strip().split():
            word = word.strip(",.*-!").lower()
    if word not  in my_dict:
        my_dict[word] = 1
    else:
        my_dict[word] += 1

sorted_items = sorted(my_dict.items(), key = lambda a: a[1], reverse=True)[:5]
print(sorted_items)
```

```python
from datetime import date

class person:
    def __init__(self, name, country, date_of_birth):
        #instance attributes
        self.name = name
        self.country = country
        self.date_of_birth = date_of_birth
    def calculate_age(self):
        #Calculates the person's current age in years.
        now = date.today()
        #calculate the initial age difference in years
        age = now.year - self.date_of_birth

        if (now.moth, now.day) < (self.date_of_birth.month, self.date_of_birth.day):
            age -= 1
        return age
    
    #example usage
person1 = person("Alice", date(1990, 6,10))
person2 = person("Bob", date(2010, 2, 23))
    #calculate age
print(f"{person1.name} is {person1.calculate_age()} years old.")
print(f"{person2.name} is {person2.calculate_age()} years old.")
```
