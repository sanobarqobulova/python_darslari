```python
#math operations.py
def add(a, b):
    return a + b
def subtract(a, b):
    return a - b
def multiply(a, b):
    return a * b
def divide(a, b):
    if b != 0:
        return a / b
    else:
        return "Cannot divide by zero!"
subtract(20,4)
multiply(4,2)
```

```python
#string_utils.py
def reverse_string(s):  #uses slicing to reverse.
     return s[:: -1] #reverse the string
def count_vowel(input_string): #counts vowels in the string.
     vowels = "aeiouAEIOU"
     count = 0 #initialize a counter
#Loop through each character in the text.
     for char in input_string:
          if char in vowels:
               count += 1
     return count
#You can pass any text inside the funcition call.
result1 = count_vowel("Hello World")     #the argument
print(f"Result: {result1}")
```

```python
import math

def calculate_area(radius):
    return math.pi * radius ** 2

def calculate_circumference(radius):
    return 2 * math.pi * radius
print(f"calculated area is: {calculate_area(12)}")
print(f"calculated circumference is: {calculate_circumference(10)}")
```

```python
#opens a file and reads it.
def read_file(file_path):
    with open(file_path, "r") as f:
        return f.read()
#opens a file and writes content.
def write_file(file_path, content):
    with open(file_path, "w") as f:
        f.write(content)
```
