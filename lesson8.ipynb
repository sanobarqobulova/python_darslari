### finally statement

```python
try:
    file = open("abc.txt")
    content = file.read()
except FileNotFoundError:
    print("file not found")
finally:
    print("file is closed")
```

```python
def check_positive(num):
    if num < 0:
        raise ValueError("the number must be positive")
    
try:
    check_positive(-5)
except ValueError as e:
    print(e)
```
