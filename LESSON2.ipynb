## 6. Count Vowels
Write a Python program that counts the number of vowels in a given string.

````python
string="this is python"
vowels="aeouiAEOUI"

count=sum(1 for char in string if char in vowels)
print(f"vowels count:{count}")#string ichidagi unli harflar soni
````

## 10. Generate Random Password
Write a Python program to generate a random password containing letters, digits, and special characters.

````python
#1 special character,one digit,one uppercase letter,length at least 8 characters
import secrets
import string

def generate_secure_password(length=12):
#combine letters, digits, and punctuation to create a secure password
     alphabet=string.ascii_letters + string.digits + string.punctuation
     #securely selects characters
     password = ''.join(secrets.choice(alphabet) for _ in range(length))
     return password     #random.choice(characters) for _ in
#generates a 12 character password
print(generate_secure_password(12))
````
