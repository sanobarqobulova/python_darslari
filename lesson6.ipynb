## nest loops

```python
matrix = [[1,2],[3,4],[5,6]]

for row in matrix:
    print(row)
```

```python
for row in matrix:
    for item in row:
     print(item)
```

```python
matrix = [[1,2],[3,4],[5,6]],[[1,2],[3,4],[5,6]]

for row in matrix:
    for row_2 in row:
        for item in row_2:
                print(item)
```

### zip funktion bir nechta sequencelarni birlashtirib bir vaqtda olishga yordam beradi

```python
fruits = ['apple','banana','pear', 'grape']
amount = [5,10,3, 7]
quantities = [12,5,9, 6]

for f, a ,q in zip(fruits, amount, quantities):
    print(f,a,q)
```
