```python
import random

def play_game():
    choices = ["tosh", "qaychi", "qog'oz"]
    
    while True:
        # Foydalanuvchi tanlovi
        user = input("Tosh, Qaychi, Qog'oz? (chiqish uchun 'exit'): ").lower()
        
        if user == 'exit':
            break
            
        if user not in choices:
            print("Noto'g'ri tanlov, qaytadan urinib ko'ring.")
            continue
            
        # Kompyuter tanlovi
        computer = random.choice(choices)
        print(f"Kompyuter: {computer}")
        
        # Natijani aniqlash
        if user == computer:
            print("Durang!")
        elif (user == "tosh" and computer == "qaychi") or \
             (user == "qaychi" and computer == "qog'oz") or \
             (user == "qog'oz" and computer == "tosh"):
            print("Siz yutdingiz! 🎉")
        else:
            print("Kompyuter yutdi! 😞")
        print("-" * 20)

play_game()
```
