** kwargs -> keyword arguments
*  is a special syntax used in function definitions to accept a variable number of keyword arguments (arguments passed as key=value pairs). The arguments are collected into a dictionary inside the function, where the keys are the argument names and the values are the argument values

```python
def introduce(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}: {value}")

introduce(name= "mike", age= 30, city = "dubai")
```

### recursive functions

```python
#5! = 120->1*2*3*4*5

def factorial(n):
    if n == 0 or n ==1:
        return 1
    else:
        return n*factorial(n-1)
print(factorial(5))    
```

```python
def roman_to_nums(s: str) -> int:
    #converts a roman to a numeral integer using iterative comparison. 
    roman_nums = {"I": 1, "V": 5, "X": 10, "L": 50, "C": 100, "D": 500, "M": 1000}
    total = 0
    n = len(s)
    i = 0
    while i < n:
 #check if the current value is less than the next value(subtractive case)       
        if i + 1 < n and roman_nums[s[i]] < roman_nums[s[i + 1]]:
            total +=roman_nums[s[i+1]] - roman_nums[s[i]]
            i +=2 #skip both characters
        else:
            total += roman_nums[s[i]]    
            i +=1 # move to the next character
            return total
print(f" 'XX' converts to: {roman_to_nums("XX")}")
```

```python
def roman_to_num(roman):
    roman_nums = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
    result = 0
    next_value = 0
    for char in reversed(roman):
        value = roman_nums[char]
        if value < next_value:
            result -= value
        else:
            result += value
        next_value = value
    return result


print(roman_to_num("MCMXCIV"))
```
