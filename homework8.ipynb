4. Write a Python program that prompts the user to input two numbers and raises a `TypeError` exception if the inputs are not numerical.

```python
user_input = input("Enter two digits:")
try: #1. validation logic
    if not user_input.isdigit():
        # manually triggering the error you want
        raise TypeError(f"{user_input} is not digit! ")
    if len(user_input) < 2:
         raise TypeError(("You must enter two digits"))
    #code that handle exceptions
except TypeError as t:
    #2. handling the triggering error
    print(f"Input Error: {t} ")
else:
     #3. success logic
     print(f"Perfect! You entered { user_input }.")
```

8. Write a Python program to find the longest words.

```python
longest_word = ""

with open('example.txt', 'r') as file:
    for line in file:
        for word in line.strip().split():
            word = word.strip('.,*-!').lower()
            if len(word) > len(longest_word):
                longest_word = word
              
print("Longest word:", longest_word)
```

20. Write a Python program to generate 26 text files named `A.txt`, `B.txt`, and so on up to `Z.txt`.

```python
import string, os
if not os.path.exists("letters"):
    os.makedirs("letters")
    for letter in string.ascii_uppercase:
        with open(letter + ".txt", "w") as f:
            f.writelines(letter)
```

21. Write a Python program to create a file where all letters of the English alphabet are listed with a specified number of letters on each line.

```python
import string
def letters_file_line(n):
    with open("words1.txt","w") as f:
        alphabet = string.ascii_uppercase
        letters = [alphabet[i:i + n] + "\n" for i in range (0, len(alphabet),n)]
        f.writelines(letters)
letters_file_line(3)
```
